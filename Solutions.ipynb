 3. Write a program to find the given number is Armstrong number or not.

```python
def armstrong(num):
    original_num = num
    length = len(str(num))
    sum = 0
    while num > 0:       
        rem = num % 10
        sum = sum + (rem ** length)
        num = num // 10
    return sum == original_num 
num = int(input("Enter the number: "))    
print("The Number is Armstrong: ", armstrong(num))
```

4. Write a program to print the Armstrong numbers between 100 to 999.

```python
def armstrong_100_999():
    lst=[]
    for i in range(100,1000):
        original_num = i
        sum = 0
        while i > 0:       
            rem = i % 10
            sum = sum + (rem ** 3)
            i = i // 10
        if original_num == sum:
            lst.append(original_num)
    return lst
print("Armstrong numbers between 100 to 999 are: ", armstrong_100_999())
```
